```python
for x in range(2,30,3):
    print ("range 2", x)
```

### Create an Iterator
To create an object/class as an iterator you have to implement the methods __iter __() and __next __() to your object.

As you have learned in the Python Classes/Objects chapter, all classes have a function called __init __(), which allows you to do some initializing when the object is being created.

The __iter __() method acts similar, you can do operations (initializing etc.), but must always return the iterator object itself.

The __next __() method also allows you to do operations, and must return the next item in the sequence.

```python
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    x = self.a
    self.a += 1
    return x

myclass = MyNumbers()
myiter = iter(myclass)

print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
```

"w" - Write - Opens a file for writing, creates the file if it does not exist

## Matplotlib Labels and Title

Create Labels for a Plot

With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

```python
plt.plot(x,y1)
plt.xlabel("time in hour")
plt.ylabel("temperature in Celsius")
plt. show()
```

Create a Title for a Plot

With Pyplot, you can use the title() function to set a title for the plot.


Position the Title

You can use the loc parameter in title() to position the title.

Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
plt.plot(x, y2)

plt.xlabel("time in hour")
plt.ylabel("temperature in Celsius")

plt.title("Temperature in Nairobi, kenya", loc="left") # settiing the title location to the left

plt.show()

```

## Set Font Properties for Title and Labels

```python
plt.plot(x,y1)

plt.title("Temperature in Ayugis", fontsize= 14, fontstyle='italic', fontfamily= 'serif',color= "y",loc= "left")
plt.show()
```
